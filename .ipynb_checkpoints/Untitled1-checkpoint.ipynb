## Clean data

```python
def tweak_countries(filepath):
    col_names = {
        'Country'                           : 'country',
        'Region'                            : 'region',
        'Population'                        : 'population',
        'Area (sq. mi.)'                    : 'area',
        'Pop. Density (per sq. mi.)'        : 'pop_density',
        'Coastline (coast/area ratio)'      : 'coastline',
        'Net migration'                     : 'migration',
        'Infant mortality (per 1000 births)': 'infant',
        'GDP ($ per capita)'                : 'gdp',
        'Literacy (%)'                      : 'litaracy',
        'Phones (per 1000)'                 : 'phones',
        'Arable (%)'                        : 'arable',  # обработваема земя
        'Crops (%)'                         : 'crops',
        'Other (%)'                         : 'other',
        'Climate'                           : 'climate',
        'Birthrate'                         : 'birthrate',
        'Deathrate'                         : 'deathrate',
        'Agriculture'                       : 'agriculture',
        'Industry'                          : 'industry',
        'Service'                           : 'service',
    }
        
    def strip_cols(df, cols):
        for col in cols:
            df[col] = df[col].str.strip()
        return df
        
    def handle_floats(df, cols):
        for col in cols:
            df[col] = df[col].str.replace(',', '.').astype(np.float64)
            # fill na with mean for the region
            mean_values = df.groupby('region')[col].transform('mean')
            df[col] = df[col].fillna(mean_values)
        return df
        
    def handle_gdp(df):
        mean_values = df.groupby('region')['gdp'].transform('mean')
        df['gdp'] = df['gdp'].fillna(mean_values)
        return df
            
    df = pd.read_csv(filepath)
    
    return (df
            .rename(columns=)
            .pipe(strip_cols, ['country', 'region'])
            .pipe(handle_floats, [
                'pop_density',
                'coastline',
                'migration',
                'infant',
                'litaracy',
                'phones',
                'arable',
                'crops',
                'other',
                'birthrate',
                'deathrate',
                'agriculture',
                'industry',
                'service',
            ])
            .pipe(handle_gdp)
            .assign(climate=lambda df_: (df_
                                         .climate.str.replace(',', '.')
                                         .astype(np.float64)
                                         .fillna(3)))  # Reformat
           )

    
df = tweak_countries('data/countries/countries of the world.csv')
print(df.shape)
df.head()
```
