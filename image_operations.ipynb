```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import skimage

from pathlib import Path
```

# Image Operations

```python
xray = skimage.io.imread('data/cara-shelton-_GpJpHnyCSw-unsplash.jpg')
xray.shape
```

```python
xray = skimage.transform.resize(xray, (300, 500))
plt.imshow(xray)
plt.show()
```

```python
# 3chanels -> 1 channel
xray = xray.mean(axis=2)
plt.imshow(xray, cmap='gray')
plt.show()
```

```python
xray.max(), xray.min()
```

```python
# Thresholding
xray[xray > 0.5] = 1
xray[xray <= 0.5] = 0
```

```python
xray
```

```python
plt.imshow(xray, cmap='gray')
plt.title('Original')
plt.show()

plt.imshow(skimage.morphology.opening(xray), cmap='gray')
plt.title('Opening')
plt.show()

plt.imshow(skimage.morphology.closing(xray), cmap='gray')
plt.title('Closing')
plt.show()
```
